```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import matplotlib.ticker as ticker

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:/Windows/Fonts/H2GTRE.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()

matplotlib.rcParams['font.family'] = [font]+current_font_list
```

```python
pd_data = pd.read_csv('../이륜차신고현황_시도별_20230309165254.csv',encoding='cp949',header=1)
pd_data['시점'] = pd_data['시점'].astype(str)
for i,v in enumerate(pd_data['시점']):
    pd_data['시점'][i] = v.split('.')[0]
```

```python
g1 = pd_data.groupby(['시군구(1)','시점'])
df_motorcycle = g1.sum()
df_motorcycle
```

```python
fig1 = plt.figure(figsize=(15,10))
a = fig1.add_subplot(2,2,1)
b = fig1.add_subplot(2,2,2)
c = fig1.add_subplot(2,2,3)
d = fig1.add_subplot(2,2,4)

l=df_motorcycle.reset_index().pivot('시군구(1)','시점','경형')
s=df_motorcycle.reset_index().pivot('시군구(1)','시점','소형')
m=df_motorcycle.reset_index().pivot('시군구(1)','시점','중형')
L=df_motorcycle.reset_index().pivot('시군구(1)','시점','대형')
sns.heatmap(l,linewidths=0.5,annot=True,fmt=".0f",linecolor='black',cmap='crest',ax=a)
sns.heatmap(s,linewidths=0.5,annot=True,fmt=".0f",linecolor='black',cmap='crest',ax=b)
sns.heatmap(m,linewidths=0.5,annot=True,fmt=".0f",linecolor='black',cmap='crest',ax=c)
sns.heatmap(L,linewidths=0.5,annot=True,fmt=".0f",linecolor='black',cmap='crest',ax=d)
a.set_title('경형')
b.set_title('소형')
c.set_title('중형')
d.set_title('대형')
```
